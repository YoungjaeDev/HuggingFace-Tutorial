```python
from transformers import pipeline
```

## **Bert 파이프라인**

```python
# pipeline에 fill-mask 및 bert 모델 선택
unmasker = pipeline('fill-mask', model='bert-base-uncased')
```

```python
unmasker("MLM and NSP is the [MASK] task of BERT")
```

## **DistilBert MLM 파이프라인**

```python
unmasker = pipeline('fill-mask', model='distilbert-base-uncased')
```

```python
unmasker("MLM and NSP is the [MASK] task of BERT")
```

```python
# Bert: 110M vs DistilBert: 60M
```

## **Albert 활용**

```python
unmasker = pipeline('fill-mask', model='albert/albert-base-v2')
```

```python
unmasker("MLM and NSP is the [MASK] task of BERT")
```
